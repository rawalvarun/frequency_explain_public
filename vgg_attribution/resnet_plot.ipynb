```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy.fftpack import dct, idct
from PIL import Image
import matplotlib.pyplot as plt
from load_cifar import load_cifar_10_data
```

## Nature Trained

```python
attr = np.load("res_nat_N_freq_attr_train.npy")
labels = np.load("res_nat_N_freq_attr_train_labels.npy")
cifar_10_dir = 'cifar-10-batches-py'

train_data, train_filenames, train_labels, test_data, test_filenames, test_labels, label_names = \
    load_cifar_10_data(cifar_10_dir)
attr.shape
```

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

def barplot(i):
    test_attr = attr[i]
    test_label = labels[i]
    plt.bar(np.arange(test_attr.shape[0]), height=test_attr, color=['blue' if l == 0 else 'red' for l in test_label])

interact(barplot, i=widgets.IntSlider(min=0, max=attr.shape[0], step=1, value=0))
```

```python
fig, ax = plt.subplots(2, 5, figsize=(18, 6))
for i in range(10):
    class_attr = attr[train_labels == i]
    if i < 5:
        ax[0, i].bar(np.arange(class_attr.shape[1]), height=class_attr.mean(axis=0))
        ax[0, i].set_title("Class " + str(i))
    else:
        ax[1, i-5].bar(np.arange(class_attr.shape[1]), height=class_attr.mean(axis=0))
        ax[1, i-5].set_title("Class " + str(i))
plt.show()

fig, ax = plt.subplots(2, 5, figsize=(18, 6))
for i in range(10):
    class_attr = attr[train_labels == i]
    if i < 5:
        ax[0, i].errorbar(np.arange(class_attr.shape[1]), class_attr.mean(axis=0), yerr=class_attr.std(axis=0))
        ax[0, i].set_title("Class " + str(i))
    else:
        ax[1, i-5].errorbar(np.arange(class_attr.shape[1]), class_attr.mean(axis=0), yerr=class_attr.std(axis=0))
        ax[1, i-5].set_title("Class " + str(i))
plt.show()

fig, ax = plt.subplots(2, 5, figsize=(18, 6))
for i in range(10):
    class_attr = attr[train_labels == i]
    if i < 5:
        ax[0, i].boxplot(class_attr)
        ax[0, i].set_title("Class " + str(i))
    else:
        ax[1, i-5].boxplot(class_attr)
        ax[1, i-5].set_title("Class " + str(i))
plt.show()
# plt.bar(np.arange(class_attr.shape[1]), height=class_attr.mean(axis=0))
# plt.show()
# plt.errorbar(np.arange(class_attr.shape[1]), class_attr.mean(axis=0), yerr=class_attr.std(axis=0))
# plt.show()
```
